```python
# install the natural language toolkit package
!pip install nltk

import nltk
nltk.download('stopwords')

# import the stopwords corpus and tokenize from the natural language toolkit package
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem import PorterStemmer
from nltk import NaiveBayesClassifier


import math

ps = PorterStemmer()
```

```python
# Splitting percentage (train vs test) - no validation
percent = 0.9
size = len(df_data_1)+1
split = math.floor(size*percent)
split
```

```python
# Generate the function to apply to the col in the DataFrame
# Find all stop words (words that do not contain important significance)-set to array
commonwords = [e.lower() for e in set(stopwords.words('english'))]
# Add 'm' & 've' to the list of stopwords
commonwords.extend(['m', 've'])
# Tokenize strings into words
tokenizer = RegexpTokenizer(r'\w+')  
# Generate list of strings excluding stop words (stem words)
string_to_list = lambda x: [ps.stem(el.lower()) for el in tokenizer.tokenize(x) if el.lower() not in commonwords]

# Ensure 'reviews.text' column is of type string
df_data_1['reviews.text'] = df_data_1['reviews.text'].apply(str)

# Apply to the DataFrame
df_data_1['Reviews'] = df_data_1['reviews.text'].apply(string_to_list)

# df_data_1 = df_data_1[:3] # Testing
# Display the first 5 rows
df_data_1.head()
```

```python
# Extract all words from all reviews
def get_all_words(dataframe):
    all_words = []
    for review in dataframe['Reviews']:
        all_words.extend(review)
    return list(set(all_words))

# Remove any duplicates
all_words = get_all_words(df_data_1)
all_words
```

```python
# Extract features based on what words in the review exists in the list of all non-stop words
def extract_features(review, all_words):
    # print(all_words)
    words = list(set(review))
    # print(words)
    features = {}
    for word in all_words:
        features[word] = (word in words)
        #print(word)
        #print(features[word])
    return features

df_data_1['Features'] = df_data_1['Reviews'].apply(lambda x:extract_features(x, get_all_words(df_data_1)))
df_data_1.tail()
# df_data_1['Features'][1]
```

```python
# Create the label that will determine if the review was positive
pos_rating_threshold = 4
df_data_1['reviews.rating'] = df_data_1['reviews.rating'].apply(float)
df_data_1['High Rating'] = df_data_1['reviews.rating'] >= pos_rating_threshold
df_data_1.head(10)
```

```python
# Extract features & label to feed to the model
#features=df_data_1['Features'].values
#label=df_data_1['High Rating'].values

features = [x for x in df_data_1['Features'].values]
# features
```

```python
label = [y for y in df_data_1['High Rating'].values]
# label
```

```python
set_all=list(zip(features,label))
# set_all
```

```python
import numpy as np
np.random.shuffle(set_all)
```

```python
# Set training at testing sets
training_set = set_all[:split]
testing_set = set_all[split:]
#set_all[1]
```

```python
classifier = NaiveBayesClassifier.train(training_set)
classifier.show_most_informative_features(10)
```

```python
nltk.classify.accuracy(classifier, testing_set)*100
```
